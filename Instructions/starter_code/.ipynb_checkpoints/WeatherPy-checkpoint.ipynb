##  Statistical Analysis
     
   What's the weather like as you approach the equator?  In my study, I took a random sample of several cities across the world and analyzed the following variables:  location, maximum temperature, humidity, cloudiness, and wind speed. I divided the cities based on those above the equator(northern hemisphere) and those below the equator (southern hemisphere).  Based on the analysis of data from the northern and southern hemisphere, there is a strong correlation between location and maximum temperature.  The following are my statistical analysis for each hemisphere as it relates to the variables:
   
Northern Hemisphere Max Temp vs. Latitude:  There is a negative correlation between temperature and latitude in the northern hemisphere.  As latitude decreases(cities that are closer to the equator), the maximum temperature increases.

Southern Hemisphere Max Temp vs. Latitude:  There is a positive correlation between temperature and latitude in the southern hemisphere.  As latitude increases(cities that are closer to the equator), the maximum temperature also increases.

Northern Hemisphere Humidity vs. Latitude Analysis:  There is no correlation between humidity and latitude in the northern hemisphere.

Southern Hemisphere Humidity vs. Latitude Analysis:  There is no correlation between humidity and latitude in the southern hemisphere.

Northern Hemisphere Cloudiness(%) vs. Latitude:  There is no correlation between cloudiness and latitude in the northern hemisphere.

Southern Hemisphere Cloudiness vs. Latitude:  There is no correlation between cloudiness and latitude in the southern hemisphere.

Northern Hemisphere Windspeed vs. Latitude:  There is no correlation between wind speed and latitude in the northern hemisphere.

Southern Hemisphere Wind Speed vs. Latitude: There is weak positive correlation between wind speed and latitude in the southern hemisphere.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from datetime import datetime
import time
from scipy.stats import linregress
import scipy.stats as st
from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file =pd.read_csv("../output_data/cities.csv")

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500) 
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#create empty data frame to hold weather information
column_names = ['City', 'Latitude', 'Longitude', 'Max Temperature', "Humidity", "Cloudiness", 'Wind Speed', 'Country', 'Date']
city_data_df = pd.DataFrame(columns = column_names)
city_data_df['City']=cities
```

```python
#print sample json request using London to see layout
url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = f"{url}appid={weather_api_key}&q="
sample_json= requests.get(query_url+'London').json()
pprint(sample_json)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = f"{url}appid={weather_api_key}&q="
lat, lon, max_temp,country =[],[],[],[] #empty list to collect weather data
humidity,cloudiness,wind_speed,date =[],[],[],[]
for index, city in enumerate(cities): #print a log of every city from the request
    try:
        response=requests.get(query_url+city).json()
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])   
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(time.strftime("%m/%d/%Y"))
        print(index,city)
    except:
        print(f'The city of {city} is missing weather information.') #use try/except to prevent code from breaking and note which city is missing data
        lat.append('NaN') #append 'NaN' for missing values so the demensions in the list will remain the same
        lon.append('NaN')
        max_temp.append('NaN')
        humidity.append(0)
        cloudiness.append('NaN')
        wind_speed.append('NaN')
        country.append('NaN')
        date.append('NaN')
city_data_df['Latitude']=lat
city_data_df['Longitude']=lon
city_data_df['Max Temperature']=max_temp
city_data_df['Humidity']=humidity
city_data_df['Cloudiness']=cloudiness
city_data_df['Wind Speed']=wind_speed
city_data_df['Country']=country
city_data_df['Date']=date
```

```python
#drop rows with missing data
city_data_clean = city_data_df.loc[(city_data_df["Latitude"]!='NaN')&(city_data_df["Longitude"]!='NaN')&(city_data_df["Max Temperature"]!='NaN')&
                              (city_data_df["Cloudiness"]!='NaN')&(city_data_df["Wind Speed"]!='NaN')]
city_data_clean_df=pd.DataFrame(city_data_clean)
city_data_clean_df.count()
```

```python
#change data from objects to numeric/date
city_data_clean_df[['Latitude','Longitude',"Cloudiness",'Wind Speed','Max Temperature']]=city_data_clean_df[['Latitude','Longitude',"Cloudiness",'Wind Speed','Max Temperature']].apply(pd.to_numeric)
city_data_clean_df['Date']= pd.to_datetime(city_data_clean_df['Date'])
kelvin_temp = city_data_clean_df['Max Temperature']
fahrenheit_temp = (kelvin_temp -273.15)*1.8+32 #convert kelvin temp to Fahrenheit
city_data_clean_df['Max Temperature'] =fahrenheit_temp 
wind_meter_per_second = city_data_clean_df['Wind Speed']
wind_miles_per_hour = (wind_meter_per_second *2.23694)
city_data_clean_df['Wind Speed'] = wind_miles_per_hour #convert windspeed from meter per sec to mph
city_data_clean_df.head()
```

## Linear Regression

```python
# Create a function to create Linear Regression plots for northern/southern hemisphere data
def scatterplot(variable_x,variable_y,x_label,y_label,title,annotate_values):
    (slope, intercept, rvalue, pvalue, stderr) = st.linregress(variable_x,variable_y)
    regression_values = variable_x *slope + intercept
    lin_reg = (f"linear regression model:  y =  {np.round(slope, decimals =3)} x +  {np.round(intercept, decimals =3)}\n r-squared: {np.round(rvalue**2,decimals=3)}")
    print(f'The correlation coefficient for {x_label} and {y_label} is {np.round(rvalue, decimals = 3)} and r-squared is {np.round(rvalue**2,decimals =3)}.')
    print(f'The linear regression model is y =  {np.round(slope, decimals =3)} x +  {np.round(intercept, decimals =3)}')
    plt.figure(figsize=(10,10))
    plt.xlabel(f'{x_label}')
    plt.ylabel(f'{y_label}')
    plt.annotate(lin_reg, annotate_values, color = 'r')
    plt.title(f"{title}")
    plt.plot(variable_x, regression_values)
    plt.scatter(variable_x,variable_y)
    plt.savefig(f"{title}.png") #save each figure as a .png file, output in the same folder as code
    plt.show()
```
